```python
#Regresion Lineal
#Arboles de decision
#random forest
import pandas as pd
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, f1_score
import pandas as pd
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score, f1_score

from sklearn.model_selection import train_test_split
import joblib
```

```python
train_df = pd.read_csv('../Data/train.csv')
test_df = pd.read_csv('../Data/test.csv')
validation_df = pd.read_csv('../Data/validation.csv')

# Separate target column
train_X = train_df.drop(columns=['Depression'])
train_Y = train_df['Depression']

test_X = test_df.drop(columns=['Depression'])
test_Y = test_df['Depression']

validation_X = validation_df.drop(columns=['Depression'])
validation_Y = validation_df['Depression']
```

Se va a realizar un experimento donde se varian hiperparametros y se busca la mejor solucion

```python
# Valores del hiperparámetro k (número de vecinos)
k_values = [3, 5, 7, 9]
weights_options = ['uniform', 'distance']
metric_options = ['euclidean', 'manhattan']

results = []
```

Se entrena el modelo con todos los posibles valores de hiperparametros y se selecciona el que tenga mejores métricas


```python
from sklearn.metrics import confusion_matrix, precision_score, recall_score

# Loop a través de las combinaciones de hiperparámetros
for k in k_values:
    for weights in weights_options:
        for metric in metric_options:
            # Inicializar el clasificador k-NN con los hiperparámetros actuales
            model = KNeighborsClassifier(n_neighbors=k, weights=weights, metric=metric)

            # Entrenar el modelo
            model.fit(train_X, train_Y)

            # Realizar predicciones en el conjunto de validación
            validation_pred = model.predict(validation_X)

            # Calcular la matriz de confusión
            cm = confusion_matrix(validation_Y, validation_pred)

            # Calcular las métricas
            accuracy = accuracy_score(validation_Y, validation_pred)
            precision = precision_score(validation_Y, validation_pred, average='weighted')
            recall = recall_score(validation_Y, validation_pred, average='weighted')
            f1 = f1_score(validation_Y, validation_pred, average='weighted')

            # Guardar los resultados
            results.append({
                'k': k,
                'weights': weights,
                'metric': metric,
                'Accuracy': accuracy,
                'Precision': precision,
                'Recall': recall,
                'F1 Score': f1,
                'Confusion Matrix': cm
            })
```

```python
# Convert the results to a DataFrame
results_df = pd.DataFrame(results)
results_df
```

```python
# Seleccionar los mejores parámetros con base en F1 Score (o Accuracy)
best_params = results_df.loc[results_df['F1 Score'].idxmax()]
best_params
```

Se entrena el modelo con los mejores parámetros.

```python
# Entrenar el mejor modelo con los hiperparámetros del mejor resultado
best_model = KNeighborsClassifier(
    n_neighbors=int(best_params['k']),
    weights=best_params['weights'],
    metric=best_params['metric']
)

best_model.fit(train_X, train_Y)
```

## Finalmente se usan los datos de "test" para validar las métricas del modelo.

```python
# Evaluate the model on the test set
test_pred = best_model.predict(test_X)
test_mse = mean_squared_error(test_Y, test_pred)
test_r2 = r2_score(test_Y, test_pred)

# Display the final model metrics on the test set
print(f"Test MSE: {test_mse}")
print(f"Test R2: {test_r2}")
```
